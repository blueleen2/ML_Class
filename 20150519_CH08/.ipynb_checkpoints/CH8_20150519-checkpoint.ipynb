```python
def standRegres(xArr, yArr):
    xMat = mat(xArr)
    yMat = mat(yArr).T
    xTx = xMat.T * xMat
   
    if linalg.det(xTx) == 0.0: ##역행렬 여부 검출
        print "This matrix is singular, cannot do inverse" 
        return
    
    ws = xTx.I * (xMat.T * yMat)
    
    return ws

ws
```

```python
def regularize(xMat):
    inMat = xMat.copy()
    inMeans = mean(inMat, 0)
    inVar = var(inMat,0)
    inMat = (inMat - inMeans) / inVar
    return inMat
```

```python
def stageWise(xArr,yArr,eps=0.02,numIt=1):  ## eps: 각 반복에서 처리하게 될 단계의 크기, numIt = 반복 횟수
    xMat = mat(xArr); yMat=mat(yArr).T  ## 입력데이터를 행렬로 변환
    yMean = mean(yMat,0)
    yMat = yMat - yMean
    xMat = regularize(xMat)   ## 정규화
    m,n=shape(xMat)
    ws = zeros((n,1)); wsTest = ws.copy(); wsMax = ws.copy()   ## Greedy Algorithm에서 사용하기 위해 두 개의 복사본을 생성
    
    for i in range(numIt):
        
        print ws.T
        lowestError = inf;   ## 오류는 초기에 무한대로 설정
        
        for j in range(n):
            
            for sign in [-1,1]:
                wsTest = ws.copy()
                wsTest[j] += eps*sign
                yTest = xMat*wsTest
                rssE = regression.rssError(yMat.A,yTest.A)   ## rssError함수 -> ((yArr-yHatArr)**2).sum() : 제곱의 오류로 계산
                
                if rssE < lowestError:
                    lowestError = rssE   ## 모든 오류와 비교하면서 가장 낮은 오류로 갱신 
                    wsMax = wsTest
                
            ws = wsMax.copy()
            #print ws
            returnMat[i,:]=ws.T
            return returnMat
```
