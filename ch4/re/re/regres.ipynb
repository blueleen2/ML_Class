```python
# 소스코드에 return 혹은 전달인자의 타입에 대해서 명시 해 줄 것
# stocGradAscent0 와 stocGradAscent1 를 비교하여 설명
# 교재에 있는 테스트를 3가지로 전부 비교해서 그림 + 평균 에러 출력

from numpy import *
%matplotlib inline

def loadDataSet():
    dataMat = []; labelMat = []
    fr = open('testSet.txt')
    for line in fr.readlines():
        lineArr = line.strip().split()
        # 여기에 왜 1이 들어가는가????
        dataMat.append([1.0, float(lineArr[0]), float(lineArr[1])])
        labelMat.append(int(lineArr[2]))
    return dataMat,labelMat

def sigmoid(inX):
    return 1.0/(1+exp(-inX))

# offline 방식으로 처리하기 위한 방법
def gradAscent(dataMatIn, classLabels):
    dataMatrix = mat(dataMatIn)             #convert to NumPy matrix
    #print dataMatrix
    labelMat = mat(classLabels).transpose() #convert to NumPy matrix
    m,n = shape(dataMatrix)
    alpha = 0.001
    maxCycles = 1500
    weights = ones((n,1))
#    print"{}, {}".format(m, n)
#    print weights
    for k in range(maxCycles):              #heavy on matrix operations\
        # 여기서 행렬곱을 이용하여 1개의 값만 전달 됨
        # 내적의 개념처럼 곱하고 더한다.
        h = sigmoid(dataMatrix*weights)     #matrix mult
        error = (labelMat - h)              #vector subtraction        
        weights = weights + alpha * dataMatrix.transpose()* error #matrix mult
#    print error
    return weights;

def plotBestFit(weights):
    import matplotlib.pyplot as plt
    dataMat,labelMat=loadDataSet()
    dataArr = array(dataMat)
    n = shape(dataArr)[0] 
    xcord1 = []; ycord1 = []
    xcord2 = []; ycord2 = []
    for i in range(n):
        if int(labelMat[i])== 1:
            xcord1.append(dataArr[i,1]); ycord1.append(dataArr[i,2])
        else:
            xcord2.append(dataArr[i,1]); ycord2.append(dataArr[i,2])
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(xcord1, ycord1, s=30, c='red', marker='s')
    ax.scatter(xcord2, ycord2, s=30, c='green')
    x = arange(-3.0, 3.0, 0.1)
    y = (-weights[0]-weights[1]*x)/weights[2]
    ax.plot(x, y)
    plt.xlabel('X1'); plt.ylabel('X2');
    plt.show()

# online 방식으로 처리하기 위한 방법
def stocGradAscent0(dataMatrix, classLabels):
    m,n = shape(dataMatrix)
    alpha = 0.01
    weights = ones(n)   #initialize to all ones  
    for i in range(m):
        h = sigmoid(sum(dataMatrix[i]*weights))
        error = classLabels[i] - h
        weights = weights + alpha * error * dataMatrix[i]
    return weights

# online 방식으로 처리하기 위한 방법
# alpha의 값을 변경해주고, random한 data를 추출하여 weight를 계산
def stocGradAscent1(dataMatrix, classLabels, numIter=150):
    m,n = shape(dataMatrix)
    weights = ones(n)   #initialize to all ones
    for j in range(numIter):
        dataIndex = range(m) # 0 ~ m-1까지의 index번호가 같은 배열을 생성(할당)
        #print dataIndex
        for i in range(m):
            alpha = 4/(1.0+j+i)+0.0001    #apha decreases with iteration, does not 
            # 동일한 data를 두번 접근하지 않게 하기 위해 uniform하게 생성
            randIndex = int(random.uniform(0,len(dataIndex)))#go to 0 because of the constant
            h = sigmoid(sum(dataMatrix[randIndex]*weights))
            error = classLabels[randIndex] - h
            weights = weights + alpha * error * dataMatrix[randIndex]
            del(dataIndex[randIndex]) # 할당된 배열을 해제
    return weights

def classifyVector(inX, weights):
    prob = sigmoid(sum(inX*weights))
    if prob > 0.5: return 1.0
    else: return 0.0

def colicTest():
    frTrain = open('horseColicTraining.txt'); frTest = open('horseColicTest.txt')
    trainingSet = []; trainingLabels = []
    for line in frTrain.readlines():
        currLine = line.strip().split('\t')
        lineArr =[]
        for i in range(21):
            lineArr.append(float(currLine[i]))
        trainingSet.append(lineArr)
        trainingLabels.append(float(currLine[21]))
    # random online
    #trainWeights = stocGradAscent1(array(trainingSet), trainingLabels, 1000) # weight를 구한다
    # offline
    trainWeights = gradAscent(trainingSet, trainingLabels) # weight를 구한다
    # online
    #trainWeights = stocGradAscent0(array(trainingSet), trainingLabels) # weight를 구한다
    
    errorCount = 0; numTestVec = 0.0
    for line in frTest.readlines():
        numTestVec += 1.0
        currLine = line.strip().split('\t')
        lineArr =[]
        for i in range(21):
            lineArr.append(float(currLine[i]))
        # random online
        #if int(classifyVector(array(lineArr), trainWeights))!= int(currLine[21]):
        # offline
        if int(classifyVector(array(lineArr), trainWeights.getA()))!= int(currLine[21]):
        # online
        #if int(classifyVector(array(lineArr), trainWeights))!= int(currLine[21]):
            errorCount += 1
    errorRate = (float(errorCount)/numTestVec)
    print "the error rate of this test is: %f" % errorRate
    return errorRate

def multiTest():
    numTests = 10; errorSum=0.0
    for k in range(numTests):
        errorSum += colicTest()
    print "after %d iterations the average error rate is: %f" % (numTests, errorSum/float(numTests))
```
