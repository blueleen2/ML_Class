###2.2 시그모이드(Sigmoid)<br><br>

로지스틱 회귀 모형의 수식은

$$
\begin{align}
ln\frac{p}{(1-p)} = a + b_1*x_1 + b_2*x_2 + b_3*x_3 + \dots
\end{align}
$$

$$
\begin{align}
p(z) = \frac{1}{1+e^{-z}}\quad(z = a + b_1*x_1 + b_2*x_2 + b_3*x_3 + \dots)
\end{align}
$$

본 교재에서는 $p(z)$를 sigmoid로 표현하였다.

수학적으로 분할하는 것이 매우 쉬운 함수로 가장 기본적으로 
$$
\begin{align}
p(z) < 0.5\,이면\quad 0
\end{align}
$$

$$
\begin{align}
p(z) > 0.5\,이면\quad 1
\end{align}
$$

###Example 2
sigmoid 함수 그리기

```python
# sigmaoid를 계산하는 함수
def sigmoid(inX):
    return 1.0/(1+exp(-inX))

def plotSigmoid(inX):
    import matplotlib.pyplot as plt # 그래프를 그릴 모듈을 import
    fig = plt.figure() # 그림을 그릴 공간을 만듬
    ax = fig.add_subplot(111) # 1X1 공간의 1번 공간에 ax 그래프를 만듬
    x = arange(-inX, inX, 0.005)
    y = sigmoid(x)
    ax.plot(x, y)
    plt.xlabel('X'); plt.ylabel('Sigmoid(X)');
    plt.show()
    
plotSigmoid(5)
plotSigmoid(50)
```

##3. 기울기(Gradient)<br><br>

$$
\begin{align}
\nabla\,f(x,\,y)\,=\,
\begin{pmatrix}
\frac{\partial f(x,\,y)}{\partial x} \\
\frac{\partial f(x,\,y)}{\partial y}
\end{pmatrix}
\end{align}
$$

위의 기호는 $\frac{\partial f(x,\,y)}{\partial x}$만큼 $x$ 방향으로,
$\frac{\partial f(x,\,y)}{\partial y}$만큼 $y$ 방향으로 이동을 의미한다.

###3.1 기울기 하강(Gradient Descent)<br><br>

함수의 기울기를 구하여 기울기가 낮은 쪽으로 계속 이동시켜서 극값에 이를 때까지 반복시키는 최적화 알고리즘으로 식은 아래와 같다.

$$
\begin{align}
x_{i+1}\,=\,x_{i}\,-\,\gamma_{i}\,\nabla f(x_{i})
\end{align}
$$

이때, $\gamma_{i}$는 이동할 거리를 조절하는 매개변수

###Example 3
기울기 하강을 이용하여 극값의 기울기를 구하기

```python
def f_prime_minimum(x):
    return 4 * x**3 - 9 * x**2

def f_function_minimum(x):
    return x**4 - 3 * x**3 + 2

def f_minimum(inX, gamma, precision):
    x_old = 0
    x_new = inX
    while abs(x_new - x_old) > precision:
        x_old = x_new
        x_new = x_old - gamma * f_prime_minimum(x_old)
    return x_new

def plotMinimum(startX):
    import matplotlib.pyplot as plt # 그래프를 그릴 모듈을 import
    fig = plt.figure() # 그림을 그릴 공간을 만듬
    ax = fig.add_subplot(111) # 1X1 공간의 1번 공간에 ax 그래프를 만듬
    x = arange(-1, 3, 0.005)
    y = f_function_minimum(x)
    ax.plot(x, y)
    minimumX = f_minimum(startX, 0.01, 0.00001)
    minimumY = f_function_minimum(minimumX)
    gradX = f_prime_minimum(minimumX)
    x1 = arange(minimumX - 1.0, minimumX + 1.0, 0.1)
    y1 = gradX*x1 + minimumY
    ax.plot(x1, y1)
    plt.xlabel('X'); plt.ylabel('Y');
    plt.show()
 
plotMinimum(6)
```

###3.2 기울기 상승(Gradient Ascent)<br><br>

함수의 기울기를 구하여 기울기가 큰 쪽으로 계속 이동시켜서 극값에 이를 때까지 반복시키는 최적화 알고리즘으로 식은 아래와 같다.

$$
\begin{align}
x_{i+1}\,=\,x_{i}\,+\,\gamma_{i}\,\nabla f(x_{i})
\end{align}
$$

이때, $\gamma_{i}$는 이동할 거리를 조절하는 매개변수

###Example 4
기울기 상승을 이용하여 극값의 기울기를 구하기

```python
def f_prime_maximum(x):
    return (-10)*(2*x**3 + 3 * x**2 - x - 1)

def f_function_maximum(x):
    return (x + 2)*(x + 1)*x*(x - 1)*(-5)

def f_maximum(inX, gamma, precision):
    x_old = 0
    x_new = inX
    while abs(x_new - x_old) > precision:
        x_old = x_new
        x_new = x_old + gamma * f_prime_maximum(x_old)
    return x_new

def plotMaximum(startX):
    import matplotlib.pyplot as plt # 그래프를 그릴 모듈을 import
    fig = plt.figure() # 그림을 그릴 공간을 만듬
    ax = fig.add_subplot(111) # 1X1 공간의 1번 공간에 ax 그래프를 만듬
    x = arange(-2, 1, 0.005)
    y = f_function_maximum(x)
    ax.plot(x, y)
    maximumX = f_maximum(startX, 0.01, 0.00001)
    maximumY = f_function_maximum(maximumX)
    gradX = f_prime_maximum(maximumX)
    x1 = arange(maximumX - 1.0, maximumX + 1.0, 0.1)
    y1 = gradX*x1 + maximumY
    ax.plot(x1, y1)
    plt.xlabel('X'); plt.ylabel('Y');
    plt.show()
 
plotMaximum(-1)
```

###3.3 약점(Weakness)<br><br>

함수의 기울기를 이용하여, 극소값 혹은 극대값을 계산해야하는데, 초기 값을 잘못 설정하면 원하는 결과를 얻을 수 없다.

```python
print "기울기 하강"
plotMinimum(6)
plotMinimum(0)
print "기울기 상승"
plotMaximum(-1)
plotMaximum(0.5)
```

##4. 가장 좋은 회귀 계수를 찾기(Best Fit Line)<br><br>

본 교재에서는 가장 좋은 회귀 계수를 찾기 위해 비선형 함수인 sigmoid와 기울기 상승을 이용하였다.

sigmoid함수는 특정 계수를 이용하여 분류를 하기 위한 예측값이 된다.

기울기를 이용한 최적화 기법은 기울기 상승/하강이 존재하는데, 상승을 사용한 이유는

회귀 후 나타나는 결과는 올바르게 판단된 확률을 측정하게 되는데 이때, 측정되는 확률은 0과 1사이의 확률로 나타내어지고, 올바르게 분류 할수록 확률값이 커지므로, 확률값이 가장 크다고 생각하는 지점을 찾을 수 있게 기울기 상승 기법을 이용한다.

따라서 Sigmoid함수의 입력값은
$$
\begin{align}
z\,=\sum^{n}_{i=0}{w_{i}x_{i}}
\end{align}
$$

이고, 새롭게 갱신되는 $w$는 기울기 상승기법을 이용하여

$$
\begin{align}
w\,:=w+\alpha\nabla_{W}f(w)
\end{align}
$$
이다.

###4.1 기울기 상승을 이용하여 가장 좋은 매개변수 찾기 (Batch 방법)

한 번에 단 모든 사례를 사용하여 가중치를 갱신하는 방법

####의사 코드
- 모두 1로 설정된 가중치를 가지고 시작
- R번 반복
    - 입력 데이터 집합의 기울기를 계산
    - $\alpha\,*\,$기울기로 가중치 벡터를 결정
    - 가중치 벡터 변환

```python
# data를 불러오는 함수
def loadDataSet():
    dataMat = []; labelMat = []  # 리스트 type으로 dataMat과 labelMat 변수를 생성
    fr = open('testSet.txt')
    for line in fr.readlines():
        # strip는 문자열의 양 끝 공백을 없애는 함수
        # split는 문자열을 쪼개는 함수, 인자값이 없으면 ' ' 을 기본으로 자른다.
        lineArr = line.strip().split()
        
        # dataMat에 3열짜리 데이터를 추가한다.
        # 세 번째 열에 1.0을 넣은 이유는 Regression 후 절편값을 계산하기 위함
        # ex) ax + by + c = 0 이라 할 때, a는 첫 번째 열로 계산되며, b는 두 번째 열, c는 세 번째 열이다.
        dataMat.append([float(lineArr[0]), float(lineArr[1]), 1.0])
        labelMat.append(int(lineArr[2]))
    return dataMat,labelMat

# 상승 Gradient를 계산하는 함수
# dataMatIn [첫 번째 요소, ..., n 번째 요소, 1.0]
# classLabel [첫 번째 그룹, ..., m 번째 그룹]
# maxCycles 매개변수를 구하기 위해서 실행되는 반복횟수
def gradAscent(dataMatIn, classLabels, maxCycles):
    dataMatrix = mat(dataMatIn)             #list 타입을 numpy의 Matrix타입으로 바꿔줌
    #print "dataMatIn type :", type(dataMatIn)
    #print "dataMatrix type :", type(dataMatrix)
    labelMat = mat(classLabels).transpose() #convert to NumPy matrix
    m,n = shape(dataMatrix) # 행과 열을 가져온다.
    #print "행의 갯수 :", m
    #print "열의 갯수 :", n
    #print dataMatrix.transpose()
    alpha = 0.001
    #maxCycles = 3000
    weights = ones((n,1))  # [행의 갯수 : Input의 열의 갯수, 열의 갯수 : 1] 만큼 1로 초기화
    for k in range(maxCycles):              # heavy on matrix operations
        # 여기서 행렬곱을 이용하여 1개의 값만 전달 됨
        # 내적의 개념처럼 곱하고 더한다.
        h = sigmoid(dataMatrix * weights)     # 각 요소에 weight 값을 곱햇서 sigmoid 함수를 호출한다.
        error = (labelMat - h)                # sigmoid를 이용해서 그룹을 나눈다.
        weights = weights + alpha * dataMatrix.transpose() * error # 각 요소의 weight 값을 조절해준다.
    return weights;

def gradDescent(dataMatIn, classLabels, maxCycles):
    dataMatrix = mat(dataMatIn)             #list 타입을 numpy의 Matrix타입으로 바꿔줌
    #print "dataMatIn type :", type(dataMatIn)
    #print "dataMatrix type :", type(dataMatrix)
    labelMat = mat(classLabels).transpose() #convert to NumPy matrix
    m,n = shape(dataMatrix) # 행과 열을 가져온다.
    #print "행의 갯수 :", m
    #print "열의 갯수 :", n
    #print dataMatrix.transpose()
    alpha = 0.001
    #maxCycles = 3000
    weights = ones((n,1))  # [행의 갯수 : Input의 열의 갯수, 열의 갯수 : 1] 만큼 1로 초기화
    for k in range(maxCycles):              # heavy on matrix operations
        # 여기서 행렬곱을 이용하여 1개의 값만 전달 됨
        # 내적의 개념처럼 곱하고 더한다.
        h = sigmoid(dataMatrix * weights)     # 각 요소에 weight 값을 곱햇서 sigmoid 함수를 호출한다.
        error = (labelMat - h)                # sigmoid를 이용해서 그룹을 나눈다.
        weights = weights - alpha * dataMatrix.transpose() * error # 각 요소의 weight 값을 조절해준다.
    return weights;

# 그래프를 그려주는 함수 
# weight type은 numpy의 Array 타입
def plotBestFit(weights):
    #print "Weights type :", type(weights)
    import matplotlib.pyplot as plt # 그래프를 그릴 모듈을 import
    # 참고 링크
    # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot
    # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter
    # http://matplotlib.org/api/markers_api.html#matplotlib.markers.MarkerStyle
    dataMat,labelMat = loadDataSet() # 표현할 data를 불러 옴
    dataArr = array(dataMat) # dataMat은 list 타입이므로 numpy의 Array 타입으로 변경
    n = shape(dataArr)[0] # dataArr의 형태를 나타내는 첫번째 요소(열의 갯수)를 가져옴
    xcord1 = []; ycord1 = [] # 첫 번째 그룹을 표현할 list를 생성
    xcord2 = []; ycord2 = [] # 두 번째 그룹을 표현할 list를 생성
    for i in range(n): # 전체 열 만큼 loop를 실행
        if int(labelMat[i])== 1: # 1번 그룹이면
            xcord1.append(dataArr[i,0]); ycord1.append(dataArr[i,1])
        else: # 1번 그룹이 아니면
            xcord2.append(dataArr[i,0]); ycord2.append(dataArr[i,1])
    fig = plt.figure() # 그림을 그릴 공간을 만듬
    ax = fig.add_subplot(111) # 1X1 공간의 1번 공간에 ax 그래프를 만듬
    
    ax.scatter(xcord1, ycord1, s=30, c='red', marker='x')
    ax.scatter(xcord2, ycord2, s=30, c='green', marker='v')
    x = arange(-3.0, 3.0, 0.1)
    y = (-weights[2]-weights[0]*x)/weights[1]
    ax.plot(x, y)
    plt.xlabel('X1'); plt.ylabel('X2');
    plt.show()
    
dataMat, labelMat = loadDataSet()
weight = gradAscent(dataMat, labelMat, 1)
plotBestFit(weight.getA())
#weight = gradAscent(dataMat, labelMat, 5)
#plotBestFit(weight.getA())
#weight = gradAscent(dataMat, labelMat, 50)
#plotBestFit(weight.getA())
weight = gradAscent(dataMat, labelMat, 500)
plotBestFit(weight.getA())

weight = gradDescent(dataMat, labelMat, 1)
plotBestFit(weight.getA())
#weight = gradDescent(dataMat, labelMat, 5)
#plotBestFit(weight.getA())
#weight = gradDescent(dataMat, labelMat, 50)
#plotBestFit(weight.getA())
weight = gradDescent(dataMat, labelMat, 500)
plotBestFit(weight.getA())
```

###4.2 확률적인 기울기 상승을 이용하여 매개변수 찾기 (Online 방법)

한 번에 단 하나의 사례를 사용하여 가중치를 갱신하는 방법

####의사 코드
- 모두 1로 설정된 가중치를 가지고 시작
- 데이터 집합 내 각각의 데이터를 반복
    - 데이터 중 하나의 기울기를 계산
    - $\alpha\,*\,$기울기로 가중치 벡터를 결정
    - 가중치 벡터 변환

```python
# online 방식으로 처리하기 위한 방법
# 상승 Gradiant를 계산하는 함수
# dataMatIn [첫 번째 요소, ..., n 번째 요소, 1.0]
# classLabel [첫 번째 그룹, ..., m 번째 그룹]
def stocGradAscent0(dataMatrix, classLabels):
    m,n = shape(dataMatrix)
    alpha = 0.01
    weights = ones(n)   #initialize to all ones  
    for i in range(m):
        h = sigmoid(sum(dataMatrix[i]*weights))
        error = classLabels[i] - h
        weights = weights + alpha * error * dataMatrix[i]
    return weights

plotBestFit(stocGradAscent0(array(dataMat), labelMat))
```

###4.3 수정된 확률적인 기울기 상승을 이용하여 매개변수 찾기 (Online 방법)

한 번에 단 하나의 사례를 사용하여 가중치를 갱신하는 방법

####의사 코드
- 모두 1로 설정된 가중치를 가지고 시작
- R번 반복
    - 데이터 집합 내 각각의 데이터를 반복
        - $\alpha$를 갱신
        - 데이터 중 랜덤하게 하나를 선택하여 기울기를 계산
        - $\alpha\,*\,$기울기로 가중치 벡터를 결정
        - 가중치 벡터 변환

```python
# online 방식으로 처리하기 위한 방법
# alpha의 값을 변경해주고, random한 data를 추출하여 weight를 계산
def stocGradAscent1(dataMatrix, classLabels, numIter=150):
    m,n = shape(dataMatrix)
    weights = ones(n)   #initialize to all ones
    for j in range(numIter):
        dataIndex = range(m) # 0 ~ m-1까지의 index번호가 같은 배열을 생성(할당)
        #print dataIndex
        for i in range(m):
            alpha = 4/(1.0+j+i)+0.0001    #apha decreases with iteration, does not 
            # 동일한 data를 두번 접근하지 않게 하기 위해 uniform하게 생성
            randIndex = int(random.uniform(0,len(dataIndex)))#go to 0 because of the constant
            h = sigmoid(sum(dataMatrix[randIndex]*weights))
            error = classLabels[randIndex] - h
            weights = weights + alpha * error * dataMatrix[randIndex]
            del(dataIndex[randIndex]) # 할당된 배열을 해제
    return weights

plotBestFit(stocGradAscent1(array(dataMat), labelMat))
```

```python
def plotConverge(weights):
    import matplotlib.pyplot as plt # 그래프를 그릴 모듈을 import
    fig = plt.figure() # 그림을 그릴 공간을 만듬
    ax0 = fig.add_subplot(311) # 3X1 공간의 1번 공간에 ax 그래프를 만듬
    ax1 = fig.add_subplot(312) # 3X1 공간의 2번 공간에 ax 그래프를 만듬
    ax2 = fig.add_subplot(313) # 3X1 공간의 3번 공간에 ax 그래프를 만듬
    x = arange(0, 20000, 1)
    x0 = weights[x, 0]
    x1 = weights[x, 1]
    x2 = weights[x, 2]
    ax0.plot(x, x0)
    ax1.plot(x, x1)
    ax2.plot(x, x2)    
    plt.xlabel('Iteration');
    plt.show()

def gradAscentWeights(dataMatIn, classLabels):
    dataMatrix = mat(dataMatIn)             #list 타입을 numpy의 Matrix타입으로 바꿔줌    
    labelMat = mat(classLabels).transpose() #convert to NumPy matrix
    m,n = shape(dataMatrix) # 행과 열을 가져온다.
    alpha = 0.01
    maxCycles = 20000
    weights = ones((n,1))  # [행의 갯수 : Input의 열의 갯수, 열의 갯수 : 1] 만큼 1로 초기화
    retWeights = empty((20000, n))
    for k in range(maxCycles):              # heavy on matrix operations
        # 여기서 행렬곱을 이용하여 1개의 값만 전달 됨
        # 내적의 개념처럼 곱하고 더한다.
        h = sigmoid(dataMatrix * weights)     # 각 요소에 weight 값을 곱햇서 sigmoid 함수를 호출한다.
        error = (labelMat - h)                # sigmoid를 이용해서 그룹을 나눈다.
        weights = weights + alpha * dataMatrix.transpose() * error # 각 요소의 weight 값을 조절해준다.
        retWeights[k] = weights.transpose()
    return retWeights;

def stocGradAscent0Weights(dataMatrix, classLabels):
    m,n = shape(dataMatrix)
    alpha = 0.01
    weights = ones(n)   #initialize to all ones
    retWeights = empty((20000, n))
    for j in range(200):
        for i in range(m):
            h = sigmoid(sum(dataMatrix[i]*weights))
            error = classLabels[i] - h
            weights = weights + alpha * error * dataMatrix[i]
            retWeights[j * m + i] = weights
    return retWeights

def stocGradAscent1Weights(dataMatrix, classLabels, numIter=200):
    m,n = shape(dataMatrix)
    weights = ones(n)   #initialize to all ones
    retWeights = empty((20000, n))
    retWeights[0] = weights
    for j in range(numIter):
        dataIndex = range(m) # 0 ~ m-1까지의 index번호가 같은 배열을 생성(할당)
        #print dataIndex
        for i in range(m):
            alpha = 4/(1.0+j+i)+0.01    #apha decreases with iteration, does not 
            # 동일한 data를 두번 접근하지 않게 하기 위해 uniform하게 생성
            randIndex = int(random.uniform(0,len(dataIndex)))#go to 0 because of the constant
            h = sigmoid(sum(dataMatrix[randIndex]*weights))
            error = classLabels[randIndex] - h
            weights = weights + alpha * error * dataMatrix[randIndex]
            retWeights[j * m + i] = weights
            del(dataIndex[randIndex]) # 할당된 배열을 해제
    return retWeights

print "4.1, x[0], x[1], x[2]"
weights = gradAscentWeights(dataMat, labelMat)
plotConverge(weights)
print "4.2, x[0], x[1], x[2]"
weights = stocGradAscent0Weights(array(dataMat), labelMat)
plotConverge(weights)
print "4.3, x[0], x[1], x[2]"
weights = stocGradAscent1Weights(array(dataMat), labelMat)
plotConverge(weights)
```

###4.4 누락된 데이터값 다루기

####처리 방법
- 사용 가능한 모든 데이터에서 속성의 평균값을 사용.
- 알려지지 않은 값은 특별한 값으로 채움.
- 사례를 무시.
- 유사한 아이템들로부터의 평균값을 사용.
- 값을 예측하기 위해 다른 기계 학습 방법을 사용.

```python
def classifyVector(inX, weights):
    prob = sigmoid(sum(inX*weights))
    if prob > 0.5: return 1.0
    else: return 0.0

def originalColicLoadDataSet():    
    fr = open('horse-colic.data.txt')    
    frTrain = open('horse-colic.data.txt'); frTest = open('horse-colic.test.txt')
    trainingSet = []; trainingLabels = []
    testSet = []; testLabels = []
    for line in frTrain.readlines():
        currLine = line.strip().split(' ')
        lineArr =[]
        for i in range(27):
            if currLine[i] != '?' :
                lineArr.append(float(currLine[i]))
            else:
                lineArr.append(float(0.0))
        trainingSet.append(lineArr)
        trainingLabels.append(float(currLine[27]) - 1.0)
        
    # random online
    trainWeights = stocGradAscent1(array(trainingSet), trainingLabels, 1000) # weight를 구한다
    errorCount = 0; numTestVec = 0.0
        
    for line in frTest.readlines():
        numTestVec += 1.0
        currLine = line.strip().split(' ')
        lineArr =[]
        for i in range(27):
            if currLine[i] != '?' :
                lineArr.append(float(currLine[i]))
            else:
                lineArr.append(float(0.0))
                
        testSet.append(lineArr)
        testLabels.append(float(currLine[27]) - 1.0)    
    return trainingSet,trainingLabels, testSet, testLabels

def colicClassfyError(trainingSet,trainingLabels, testSet, testLabels):
    trainWeights = stocGradAscent1(array(trainingSet), trainingLabels, 1000) # weight를 구한다
    errorCount = 0
    numTestVec = shape(array(testSet))[0]
    for k in range(numTestVec):
        if int(classifyVector(array(testSet[k]), trainWeights))!= int(testLabels[k]):
            errorCount += 1
    errorRate = (float(errorCount)/numTestVec)
    print "the error rate of this test is: %f" % errorRate
    return errorRate

def originalColicMultiTest():
    trainingSet,trainingLabels, testSet, testLabels = originalColicLoadDataSet()
    numTests = 10; errorSum=0.0
    for k in range(numTests):
        errorSum += colicClassfyError(trainingSet,trainingLabels, testSet, testLabels)
    print "after %d iterations the average error rate is: %f" % (numTests, errorSum/float(numTests))
    
originalColicMultiTest()
```
